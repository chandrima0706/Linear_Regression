```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
```

```python
cherr = pd.read_excel('cherapunji_JJAS_yearlymean.xlsx') #import data
gang = pd.read_excel('gangtok_JJAS_yearlymean.xlsx')
imp = pd.read_excel('imphal_JJAS_yearlymean.xlsx')
shil = pd.read_excel('shilong_JJAS_yearlymean.xlsx')
```

```python
plt.plot(cherr.iloc[:,4]) #see how the maxT looks like for each stations
# plt.plot(gang.iloc[:,4])
# plt.plot(imp.iloc[:,4])
# plt.plot(shil.iloc[:,4])
```

#### Shilong

```python
x = shil['YEAR']
y = shil['..MAX']
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
```

```python
print (r_value,p_value)
```

```python
plt.plot(x, y,'o')
plt.plot(x, intercept + slope * x, '-',label = 'Y='+str(slope) +'Xx' + '+' + str(intercept))
plt.legend()
plt.xlabel('Year')
plt.ylabel('MaxT')
# plt.savefig('linearreg_ensorain_input.png')
# plt.show()
plt.title('Shilong MaximumT')
```
